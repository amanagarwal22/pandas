# 1. Pie Charts

```python
plt.pie(y, labels=y.index)
```

```python
plt.pie(y, labels=y.index, autopct="%0.2f%%")
plt.show()
```

### Calculating Pie Chart For Any Batsman

```python
def runs(a):
    x=delivery[delivery['batsman']==a]
    y=x['batsman_runs'].value_counts()
    plt.pie(y, labels=y.index, autopct="%0.2f%%")
    plt.show()
```

```python
runs('RG Sharma')
```

# LINE CHARTS

```python
# Step 1 Merge
mega=match.merge(delivery,left_on='id',right_on='match_id')
```

```python
batsman_data=mega[mega['batsman']=='V Kohli']
final=batsman_data.groupby('season').sum()['batsman_runs']
```

```python
plt.plot(final.index,final.values ,color='red',marker='D')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.title("Batsman Season by Season Performance")
```

### Runs Comparison between two batsman

```python
def compare(a,b):
    mega=match.merge(delivery,left_on='id',right_on='match_id')
    b1=mega[mega['batsman']==a]
    b2=mega[mega['batsman']==b]
    final1=b1.groupby('season').sum()['batsman_runs']
    final2=b2.groupby('season').sum()['batsman_runs']
    plt.plot(final1.index,final1.values ,color='green',marker='.',label=a)
    plt.plot(final2.index,final2.values ,color='red',marker='D',label=b)
    plt.xlabel('Season')
    plt.ylabel('Runs Scored')
    plt.title("Batsman Season by Seadon Performance")
    plt.legend()
```

```python
compare('V Kohli','MS Dhoni')
```
