```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Distribution Graphs
# ------> 1.Distplot
# ------> 2.Jointplot
# ------> 3.Pairplot
# ------> 4.Rugplot

# Categorical Plots
# ------> 1.Barplot
# ------> 2.Countplot
# ------> 3.Boxplot
# ------> 4.Violinplot

# Matrix Plot
# ------> 1. Heatmap
```

# DISTRIBUTION GRAPHS

### 1.Distplot

```python
sns.distplot(tips['total_bill']) #Use in finding numerical data distribution
```

### 2.Jointplot

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

### 3.Pairplot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

### 4.Rugplot

```python
sns.rugplot(tips['total_bill'])
```

# CATEGORICAL GRAPH

### 1.Barplot

```python
sns.barplot(x='day',y='tip',data=tips)     #internally calculating the mean
```

```python
sns.barplot(x='day',y='tip',data=tips, hue='sex')    #hue parameter gives the additional piece of information
```

```python
import numpy as np
sns.barplot(x='day',y='tip',data=tips, hue='sex',estimator=np.median)
```

### 2.Countplot

```python
sns.countplot(tips['smoker'])        #counting the frequency of a particular paramter
```

```python
sns.countplot(tips['smoker'],data=tips,hue='sex')
```

### 3.Boxplot

```python
sns.boxplot('total_bill',data=tips)               #used in only one numerical column
```

```python
sns.boxplot(x=tips['sex'],y=tips['tip']) 
```

### 4.Violinplot

```python
sns.violinplot(x=tips['sex'],y=tips['tip'])
```

```python
sns.violinplot(x=tips['day'],y=tips['tip'])
```
